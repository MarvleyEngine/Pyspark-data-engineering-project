### WHEN - OTHERWISE

#### SCENARIO 1

```python
df = df.withColumn("cat_flag", when(col("Item_Type")== "Meat", "Non Veg").otherwise("Veg"))
```

####  timestamp 3:38:29

```python
df.display()
```

```python
df = df.withColumn(
    "2nd_flag",
    when((col("cat_flag") == "Veg") & (col("Item_MRP") < 100), "Veg inexpensive")
    .otherwise("Veg expensive")
)
```

```python
df.withColumn(
    "3rd_flag",
    when((col("cat_flag") == "Veg") & (col("Item_MRP") < 100), "Veg inexpensive")
    .when((col("cat_flag") == "Veg") & (col("Item_MRP") > 100), "Very expensive")  # no backslash here
    .otherwise("Vve")
).display()
```

```python
%sql
SELECT 
  Item_Identifier,
  Item_Visibility,
  SUM(Item_Visibility) OVER (PARTITION BY Item_Identifier ORDER BY Item_Visibility) AS running_visibility
FROM default.newdata_delta;
```

```python
%sql
SELECT 
  Item_Identifier,
  Item_Fat_Content,
  AVG(Item_Visibility) OVER (PARTITION BY Item_Identifier ORDER BY Item_Identifier) AS avg_visibility
FROM default.newdata_delta;
```
